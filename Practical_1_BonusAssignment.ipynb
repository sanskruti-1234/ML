```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
import plotly.express as px
```

```python
try:
    df = pd.read_csv('Walmart_Sales.csv')
except FileNotFoundError:
    print("Error: CSV file not found. Please provide the correct path.")
    exit()

print(df.head())
print(df.info())
print(df.describe())
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['DayOfWeek'] = df['Date'].dt.dayofweek
print(df.isnull().sum())
```

```python
imputer_num = SimpleImputer(strategy='mean')
for col in df.select_dtypes(include=np.number).columns:
    df[col] = imputer_num.fit_transform(df[[col]])
```

```python
imputer_cat = SimpleImputer(strategy='most_frequent')
# Exclude 'Date' column from imputation
for col in df.select_dtypes(exclude=np.number).columns:
    if col != 'Date':  # Add this condition to skip 'Date' column
        df[col] = imputer_cat.fit_transform(df[[col]])
```

```python
object_cols = df.select_dtypes(include=['object']).columns

for col in object_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
features = ['Store', 'DayOfWeek', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'IsHoliday', 'Weekly_Sales']  # Include relevant features
target = 'Weekly_Sales'
```

```python
numerical_features = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment'] # Select numerical features
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
fig = px.scatter(df, x="Temperature", y="Weekly_Sales", color="Store", title="Temperature vs. Weekly Sales")
fig.show()
```

```python
fig = px.bar(df, x="Store", y="Weekly_Sales", title="Average Weekly Sales by Store")
fig.show()
```

```python
fig = px.line(df, x="Date", y="Weekly_Sales", title="Weekly Sales Over Time")
fig.show()
```
